### Exercício 1: Similaridade por Cosseno.

Com base na metodologia apresentada na Prática 2, usando o corpus de Machado, Construa vetorizações, binárias, por frequência e por tf-idf. Compare a ordenação dos resultados para uma consulta qualquer, com pelo menos 3 palavras. Para a ordenação dos resultados, use o cosseno entre os vetores da consulta e dos documentos.

<hr>

```python
textos = [machado.raw(id) for id in machado.fileids()]
len(textos)
```

```python
tok_textos = [WordPunctTokenizer().tokenize(t.lower()) for t in textos]
vocab = set([])
for t in tok_textos:
    vocab = vocab.union(set(t))
print(len(vocab))
```

Temos 246 textos e 66169 tokens únicos no corpus.

Usarei a mesma consulta da prática 2: "Lamartine elixir harpa".

```python
q = "Lamartine elixir harpa"
qt = WordPunctTokenizer().tokenize(q.lower())
```

Construindo as vetorizações binárias, por frequência e por tf-idf para os documentos

```python
T = nltk.TextCollection(tok_textos)
binary_matrix = np.empty((len(tok_textos), len(qt)))
tf_matrix = np.empty((len(tok_textos), len(qt)))
tfidf_matrix = np.empty((len(tok_textos), len(qt)))

for j, w in enumerate(qt):
    for i, d in enumerate(tok_textos):
        tf_matrix[i,j] = T.tf(w, d)
        tfidf_matrix[i,j] = T.tf_idf(w, d)
        if tf_matrix[i,j] > 0:
            binary_matrix[i,j] = 1
```

Construindo as vetorizações binárias, por frequência e por tf-idf para a busca considerando como se ela fosse um documento

```python
q_binary = np.array([1,1,1])
q_tf = np.array([T.tf(w, qt) for w in qt])
q_tfidf = np.array([T.tf_idf(w, qt) for w in qt])
```

Função que calcula a similaridade do cosseno entre o termo de busca e todos os documentos e ordena o resultado

```python
def cosseno(q, doc_matrix):
    ret = dict()
    norm_q = np.linalg.norm(q)
    for i in range(doc_matrix.shape[0]):
        norm_doc = np.linalg.norm(doc_matrix[i])
        if norm_doc > 0:
            ret[i] = (np.dot(q, doc_matrix[i]) / (norm_q * norm_doc))
        else:
            ret[i] = 0
    return sorted(ret.items(), key = lambda x: x[1], reverse = True)
```

Utilizando esta função

```python
cosseno(q_binary, binary_matrix)
```

```python
cosseno(q_tf, tf_matrix)
```

```python
cosseno(q_tfidf, tfidf_matrix)
```

As consultas com as três vetorizações tiveram um resultado similar visto que a ordenação foi parecida. O top 5 dos documentos considerados relevantes é o mesmo nas três estratégias de vetorização.

### Exercício 2: Explorando o conceito de similaridade

Construa uma matriz de similaridade **(cossenos entre vetores de frequência)** entre todos os textos de machado. 
Construa um histograma com os valores de similaridade, excluindo os 0s e 1s. Interprete.

<hr>

Calculando a matriz dos vetores de frequência

```python
mat = CountVectorizer().fit_transform(textos)
mat
```

Computando a similaridade do cosseno entre as linhas desta matriz

```python
cos = cosine_similarity(mat)
cos.shape
```

Plotando o histograma

```python
values = sum(cos.tolist(), [])
values = [round(v, 10) for v in values]
values = [v for v in values if v > 0 and v < 1]
```

```python
plt.hist(values, bins = 50)
plt.ylabel("Frequência")
plt.xlabel("Similaridade do Cosseno entre os vetores tf")
plt.show()
```

Vemos no histograma que grande parte das similaridades entre os vetores de frequência dos documentos está entre 0,7 e 0,9.

Refazendo agora para o tf-idf

```python
mat = TfidfVectorizer().fit_transform(textos)
cos = cosine_similarity(mat)
values = sum(cos.tolist(), [])
values = [round(v, 10) for v in values]
values = [v for v in values if v > 0 and v < 1]
plt.hist(values, bins = 50)
plt.ylabel("Frequência")
plt.xlabel("Similaridade do Cosseno entre os vetores tf-idf")
plt.show()
```

Existe uma diferença significativa entre o histograma do tf-idf e o do tf. Os valores da similaridade do cosseno considerando a vetorização tf-idf estão mais concentrados em valores mais baixos em comparação com a vetorização tf.

### Exercício 3: Usando o Gensim
Instale o Gensim, se ainda não o tiver. Seguindo o início deste tutorial: https://radimrehurek.com/gensim/tut2.html, construa um corpus usando as classes do Gensim, mas com os textos de Machado.

<hr>

Criando o corpus (seguindo o tutorial e a aula prática 2)

Tokenizando, removendo stopwords e palavras que só apareceram uma vez.

```python
sw = stopwords.words('portuguese') + list(punctuation) + ["\x97", "\x93", "...", "--"] # adicionei tokens que haviam aparecido no lsi

textos_limpos = []
c = Counter([])
for texto in textos:
    tlimpo = [token.lower() for token in WordPunctTokenizer().tokenize(texto) if token.lower() not in sw]
    c.update(tlimpo)
    textos_limpos.append(tlimpo)
textos_limpos = [[token for token in t if c[token] > 1] for t in textos_limpos]

dicionario = corpora.Dictionary(textos_limpos)

corpus = [dicionario.doc2bow(d) for d in textos_limpos]
```
